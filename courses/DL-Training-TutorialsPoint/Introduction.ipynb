## MNIST Example

```python
import numpy as np
import torch
import torchvision
import matplotlib.pyplot as plt
from torch import nn
from torchvision import datasets, transforms
from collections import OrderedDict
```

```python
transform = transforms.Compose([transforms.ToTensor(),transforms.Normalize((0.5,),(0.5))])
```

```python
mnist_train_dataset = datasets.MNIST('mnist_train_data',download=True,train=True,transform=transform)
```

PyTorch DataLoader - a utility class that works with a Dataset object to provide an efficient and easy way to load and preprocess data in batches during model training or inference. It handles shuffling, batching, and parallel loading of data, making it easier to feed data into a model for training or evaluation.

```python
from torch.utils.data import DataLoader
mnist_train_data_loader = DataLoader(mnist_train_dataset,batch_size=64,shuffle=True)
```

## Build Simple Neural Network

```python
import torch.nn.functional as F

class MNISTClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.hidden_layer = nn.Linear(784,256)
        self.output_layer = nn.Linear(256,10)
    
    def forward(self,x):
        x = self.hidden_layer(x)
        x = F.relu(x)
        x = self.output_layer(x)
        x = F.softmax(x,dim=1)
        return x
```

```python
model = MNISTClassifier()
```

```python
model.hidden_layer
```

```python
model.output_layer
```

```python
model.hidden_layer.bias
```

```python
model.hidden_layer.weight.data.fill_(0)
```

```python
criterion = nn.CrossEntropyLoss()
```

```python
train_images_batch = train_images.view(train_images.shape[0],-1)
```

```python
train_images_batch.shape
```

```python
output_predictions = model(train_images_batch)
```

```python
loss = criterion(output_predictions,train_labels)
```

```python
optimizer = torch.optim.SGD(model.parameters(),lr=0.05)
```

```python
epochs = 8
for e in range(epochs):
    train_loss = 0
    for train_images,train_labels in mnist_train_data_loader:
        images = train_images.view(train_images.shape[0],-1)
        optimizer.zero_grad()
        output_predictions = model(images)
        loss = criterion(output_predictions,train_labels)
        loss.backward()
        optimizer.step()
        train_loss+=loss.item()
    print("Training Loss for epoch =",e,"is =",train_loss)
```

### Pipeline

```python
class MnistClassifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.hidden_layer1=nn.Linear(784,392)
        self.hidden_layer2=nn.Linear(392,256)
        self.output_layer=nn.Linear(256,10)
    
    def forward(self,x):
        x = F.relu(self.hidden_layer1(x))
        x = F.relu(self.hidden_layer2(x))
        x = F.softmax(self.output_layer(x),dim=1)
        return x

class MNISTPipeline:
    def __init__(self):
        self.init_data()
        self.model = MnistClassifier()
        self.criterion = nn.CrossEntropyLoss()
        self.optimizer = torch.optim.Adam(self.model.parameters(),lr=0.0005)

    def init_data(self):
        transform = transforms.Compose([transforms.ToTensor(),transforms.Normalize((0.5),(0.5))])
        train_mnist_data = datasets.MNIST('mnist_train_data',download=True,train=True,transform=transform)
        self.train_data = train_mnist_data
        self.train_dataloader = DataLoader(train_mnist_data,batch_size=64,shuffle=True)

    def fit(self,epochs=8):
        for e in range(epochs):
            train_loss = 0
            for train_images, train_labels in self.train_dataloader:
                images = train_images.view(train_images.shape[0],-1)
                self.optimizer.zero_grad()
                output_predictions = self.model(images)
                loss = self.criterion(output_predictions,train_labels)
                loss.backward()
                self.optimizer.step()
                train_loss+=loss
            print(f"Epoch {e}, Loss={train_loss}, Average Loss={train_loss/len(self.train_dataloader)}")

pipeline = MNISTPipeline()
```

```python
pipeline.fit()
```

## Performing Inference

```python
images,labels = next(iter(mnist_train_data_loader))
random_test_image,random_test_label = images[15], labels[15]
plt.imshow(random_test_image[0],cmap='gray')
random_test_image = random_test_image.view(1,-1)
with torch.no_grad():
    output = pipeline.model(random_test_image)
print(output,"\nPrediction :",np.argmax(output))
```
